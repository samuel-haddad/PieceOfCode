## Currency formatting cleanup with REGEX

```
#-------------------- Currency formatting cleanup with REGEX --------------------#

# libs
from pyspark.sql.functions import when, col, regexp_replace


# dataframe from dict with diff values (BRL)
data = [{'id':1, 'value':'R$1000'},
        {'id':2, 'value':'R$ 2000'},
        {'id':3, 'value':'3000.55'},
        {'id':4, 'value':'4000,00'},
        {'id':5, 'value':'R$5000,50'},
        {'id':6, 'value':'R$ 6000,64'},
        {'id':7, 'value':'r$ 7000,21'},
        {'id':8, 'value':'r$8000,8'},
        {'id':9, 'value':'r$ 9000.5'},
        {'id':10, 'value':'r$10000.37'},
        {'id':11, 'value':'11000.98'}
       ]

# create dataframe
source = spark.createDataFrame(data=data)


df = (source
      .select(*source, regexp_replace("value","\\,",".").alias("value_silver")) # replace comma with dot
      .select(*source,
              regexp_replace("value_silver","[^0-9.]","").cast("float").alias("value_gold") # just numbers and dot
             )
     )

# show
display(df)
```
