# Combinatorial Analysis
Documentation: https://docs.python.org/3/library/itertools.html

```
#+------------------------------+ COMBINATIONS +------------------------------+

from pyspark.sql.functions import col, when
from pyspark.sql.types import StructType, StructField, StringType
from math import comb
import itertools

# list of items
list_items = [
    'label1'
    ,'label2'
    ,'label13'
    ,'label14'
    ,'label15'
    ,'label16'
]

# items combination
combinations = []
for L in range(len(list_items) + 1): # first is 0
    for subset in itertools.combinations(list_items, L): # combs of all lenghs
        combinations.append(list(subset))

combinations = combinations[1:]

# labeling combinations
comb_list = []
for i, a in zip(combinations, range(len(combinations))):
    comb_list.append(((i), f'comb_{a}'))
    
# dataframe with all combinations
data = comb_list
schema = StructType([
    StructField('list', StringType(), True),
    StructField('comb', StringType(), True),
])

df = spark.createDataFrame(data=data, schema=schema)

# check: are all combs here?
C = comb(6, 1)+comb(6, 2)+comb(6, 3)+comb(6, 4)+comb(6, 5)+comb(6, 6)
c = df.count()

# show
print(f'Number of combinations: {C} | Dataframe lengh {c}')
display(df)
```
